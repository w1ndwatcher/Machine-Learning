## Imputing Null values

```python
# columns with missing values contain categorical data so impute with mode
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="most_frequent")
na_columns = ['passenger_count','RatecodeID','store_and_fwd_flag','congestion_surcharge','Airport_fee']
train[na_columns] = imputer.fit_transform(train[na_columns])

test[na_columns] = imputer.transform(test[na_columns])
```

```python
# treat columns with few unique values as categorical data - convert to object for encoding
# train
train['passenger_count'] = train['passenger_count'].astype('int64')
train['improvement_surcharge'] = train['improvement_surcharge'].astype('object')
train['congestion_surcharge'] = train['congestion_surcharge'].astype('object')
train['PULocationID'] = train['PULocationID'].astype('object')
train['DOLocationID'] = train['DOLocationID'].astype('object')
train['VendorID'] = train['VendorID'].astype('object')

# test
test['passenger_count'] = test['passenger_count'].astype('int64')
test['improvement_surcharge'] = test['improvement_surcharge'].astype('object')
test['congestion_surcharge'] = test['congestion_surcharge'].astype('object')
test['PULocationID'] = test['PULocationID'].astype('object')
test['DOLocationID'] = test['DOLocationID'].astype('object')
test['VendorID'] = test['VendorID'].astype('object')
```

```python
train.dtypes
```

Feature engineering

```python
# duration of the trip in seconds
train['tpep_pickup_datetime'] = pd.to_datetime(train['tpep_pickup_datetime'])
train['tpep_dropoff_datetime'] = pd.to_datetime(train['tpep_dropoff_datetime'])
train['duration'] = (train['tpep_dropoff_datetime'] - train['tpep_pickup_datetime']).dt.total_seconds()

# pickup year, month and day (mon,tue etc.)
train['pickup_year'] = train['tpep_pickup_datetime'].dt.year
train['pickup_month'] = train['tpep_pickup_datetime'].dt.month
train['pickup_day'] = train['tpep_pickup_datetime'].dt.dayofweek
train.head()
```

```python
# drop the datetime columns, difficult to use as it is in model
train = train.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime'], axis=1)
```

```python
# same set of operations in test dataset
test['tpep_pickup_datetime'] = pd.to_datetime(test['tpep_pickup_datetime'])
test['tpep_dropoff_datetime'] = pd.to_datetime(test['tpep_dropoff_datetime'])
test['duration'] = (test['tpep_dropoff_datetime'] - test['tpep_pickup_datetime']).dt.total_seconds()
test['pickup_year'] = test['tpep_pickup_datetime'].dt.year
test['pickup_month'] = test['tpep_pickup_datetime'].dt.month
test['pickup_day'] = test['tpep_pickup_datetime'].dt.dayofweek
```

```python
test = test.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime'], axis=1)
```

## Handling categorical data

```python
X_train.dtypes
```

```python
# one hot encoding
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

ohe = OneHotEncoder(handle_unknown='ignore')
ohe_columns = ['VendorID','RatecodeID', 'payment_type', 'improvement_surcharge', 'congestion_surcharge', 'Airport_fee', 'pickup_day']

enc_train = pd.DataFrame(ohe.fit_transform(X_train[ohe_columns]).toarray())
enc_test = pd.DataFrame(ohe.transform(test[ohe_columns]).toarray()) 

X_train_encoded = X_train.join(enc_train) 
test_encoded = test.join(enc_test)
```

```python
X_train_encoded.head()
```

```python
X_train_encoded.drop(columns=ohe_columns, inplace=True)
test_encoded.drop(columns=ohe_columns, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

X_train_encoded['store_and_fwd_flag_encoded'] = label_encoder.fit_transform(X_train_encoded['store_and_fwd_flag'])
test_encoded['store_and_fwd_flag_encoded'] = label_encoder.transform(test_encoded['store_and_fwd_flag'])

X_train_encoded = X_train_encoded.drop(['store_and_fwd_flag'], axis=1)
test_encoded = test_encoded.drop(['store_and_fwd_flag'], axis=1)
```

```python
X_train_encoded.columns = X_train_encoded.columns.astype(str)
test_encoded.columns = test_encoded.columns.astype(str)
# needed before scaling
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_encoded) 
X_test_scaled = scaler.transform(test_encoded)
```
